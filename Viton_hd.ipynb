```python
from PIL import Image
import matplotlib.pyplot as plt

person = Image.open("datasets/test/image/07573_00.jpg")
cloth = Image.open("datasets/test/cloth/11791_00.jpg")
output = Image.open("results/testiii/07573_11791_00.jpg")

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(person)
plt.title("👤 Person Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(cloth)
plt.title("👕 Cloth Image")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(output)
plt.title("🧪 Try-On Result")
plt.axis("off")

plt.tight_layout()
plt.show()
```
